```python
def ten_sum():
    x = int(input())
    y = int(input())
    for i in range(x):
        count = 0
        table = []
        if i >= 1:
            y = int(input())
        for j in range(y):
            z = input()
            a = z.split(" ")
            a = [int(i) for i in a]
            table.append(a)
        for i in range(len(table)):
            checklist = [] # แนวนอน
            for j in range(len(table)):
                checklist.append(table[i][j])
                if sum(checklist) > 10:
                    checklist.pop(0)
                if sum(checklist) == 10:
                    count += 1
                # จบแนวนอน
        for i in range(len(table)):
            checklist2 = [] # แนวตั้ง
            for j in range(len(table)):
                checklist2.append(table[j][i])
                if sum(checklist2) > 10:
                    checklist2.pop(0)
                if sum(checklist2) == 10:
                    count += 1
                # จบแนวตั้ง
        return count
        
if __name__ == '__main__':
    count = ten_sum()
    print(str(count) + " ชุด")
```

```python
array = [[1,3,5,7],[2,4,8,2],[6,3,1,1],[2,3,5,6]]
def countSummation(array: list) -> int:
    result, count = 0, 0
    n = len(array)

    for i in range(0, n):
        for j in range(0, n):

            # counts the number of summation of vertical numbers
            for k in range(j + 1, n):
                result += (array[k-1][i] + array[k][i]) if (k == j + 1) else (array[k][i])
                if (result >= 10 or k == n - 1):
                    count += 1 if (result == 10) else 0
                    result = 0

            # counts the number of summation of horizontal numbers
            for k in range(j + 1, n):
                result += (array[i][k-1] + array[i][k]) if (k == j + 1) else (array[i][k])
                if (result >= 10 or k == n - 1):
                    count += 1 if (result == 10) else 0
                    result = 0

    return count

def userInput() -> list:
    dimensions = input()
    array = list()

    for _ in range(0, int(dimensions)):
        inputList = input().split()
        mapObject = map(int, inputList)
        array.append(list(mapObject))

    return array

if __name__ == '__main__':
    print(countSummation(userInput()))
```
